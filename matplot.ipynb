## Matplot 라이브러리 사용하기


```
import matplotlib.pyplot as plt
```

### 기본 그래프 그리기
- matplotlib.pyplot 모듈의 plot()함수에 하나의 숫자 리스트를 입력하면 plot()함수의 리스트 값을 y값들이라 가정하고 x값을 자동으로 만들어준다.

```
plt.plot([1, 2, 3, 4])
plt.show()
```

```
# plot([x], [y])
plt.plot([1, 2, 3, 4], [1, 4, 8, 16])
plt.show()
```

### Style 지정하기
- x, y값 인자에 대해 선의 색상과 형태를 지정하는 포맷을 세번째 인자에 입력할 수 있다.
- 'ro'는 빨간색(red)의 원형(o)마크를 의미한다.
- 예를 들어 'b-'는 파란색(blue)의 실선(-)을 의미한다.
- axis()함수를 통해 축의 범위([xmin, xmax, ymin, ymax])를 지정해준다.

```
plt.plot([1, 2, 3, 4], [1, 4, 8, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

### 여러개의 그래프 그리기
- Matplotlib는 기본적으로 Numpy Array를 사용하는데 명시적으로 Array를 사용하지 않아도 내부적으로 Array로 변환시켜준다.

```
import numpy as np
```

```
t = np.arange(0., 5., 0.2) # 200ms 간격으로 균일하게 샘플링된 시간

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### 축(axis) 레이블 설정하기
- Matplotlib의 모듈인 xlabel, ylabel을 사용하면 축에 레이블을 표시할 수 있다.

```
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('value of x')
plt.ylabel('value of y')
plt.show()
```

### 범례 표시하기
- 범례(Legend)는 그래프에 데이터의 종류를 표시하기 위한 텍스트다.
- legend()함수를 사용한다.

```
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.show()
```

- legend()함수의 loc 속성을 이용하면 범례의 위치를 지정해줄 수 있다
- 디폴트의 경우엔 최적의 위치에 나타내준다.

```
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='lower right')
plt.show()
```

- ncol 속성을 이용해 열 개수를 지정해 줄 수도 있다.

```
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label='Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc='best', ncol=2)
plt.show()
```

### 마커 사용하기
- 'bo--' 파란색 원형 마커와 점선을 의미한다.
- 'bo-' 파란색 원형 마커와 실선을 의미한다.

```
plt.plot([1, 2, 3, 4], [2, 3, 6, 15], 'bo--')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

### 그래프 색상 지정하기
- color 속성을 이용해면 그래프의 색을 지정할 수 있다.

```
plt.plot([1, 2, 3, 4], [2.0, 2.5, 3.3, 4.5], color='springgreen')
plt.plot([1, 2, 3, 4], [2.0, 2.8, 4.3, 6.5], color='violet')
plt.plot([1, 2, 3, 4], [2.0, 3.0, 5.0, 10.0], color='dodgerblue')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.show()
```

### 그래프 영역 채우기
- 그래프의 특정 영역을 색상으로 채울 수 있다.
- fill_between() : 두 수평 방향의 곡선 사이를 채웁니다.
- fill_betweenx() : 두 수직 뱡향의 곡선 사이를 채웁니다.
- fill() : 다각형 영역을 채웁니다.

```
x = [1, 2, 3, 4]
y = [2, 3, 5, 15]

plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_betweenx(y[2:4], x[2:4], color='pink', alpha=0.5)
plt.show()
```

```
plt.plot(x, y)
plt.fill_between(x[1:3], y[1:3], color='skyblue')
plt.show()
```

- 두 그래프 사이 영역 채우기

```
y1 = [1, 2, 4, 8]

plt.plot(x, y)
plt.plot(x, y1)
plt.fill_between(x[2:4], y[2:4], y1[2:4], color='grey')
plt.show()
```

### 막대 그래프(Bar Graph) 그리기
- 막대 그래프는 범주가 있는 데이터 값을 막대로 표현해준다.


```
x = np.arange(3)

years = ['2018', '2019', '2020']
values = [100, 400, 50]

plt.bar(x, values)
plt.xticks(x, years) # xticks : x축 눈금 레이블
plt.show()
```

### 산점도 그리기
- 산점도는 두 변수의 상관 관계를 직교 좌표계의 평면에 점으로 표현하는 그래프이다.


- rand()함수를 이용해 [0,1) 범위의 난수를 50개를 생성해서 x, y에 입력한다.

```
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)

plt.scatter(x, y)
plt.show()
```

- scatter의 s, c 파라미터는 마커의 크기와 색상을 지정한다.

```
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors)
plt.show()
```

### 투명도와 컬러맵 설정하기
- alpha : 마커의 투명도 지정(0~1)
- cmap : 컬러맵에 해당하는 문자열 지정

```
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()
plt.show()
```

### 히트맵 그리기
- 히트맵은 다양한 값을 갖는 숫자 데이터를 열분포 형태와 같이 색상을 이용해서 시각화해준다.

```
arr = np.random.standard_normal((30, 40))

plt.matshow(arr)

plt.show()
```

```
plt.matshow(arr)
plt.colorbar()
plt.show()
```

### 여러개의 그래프 그리기
- subplot() 함수를 이용해서 여러개의 그래프를 그릴 수 있다.

```
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)   # nrows=2, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')

plt.subplot(2, 1, 2)  # nrows=2, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')

plt.tight_layout()
plt.show()
```

### 객체 지향 인터페이스
- subplots() 메소드를 사용할 수 있다.

```
fig, ax = plt.subplots(2, 2)
plt.show()
```

```
x = np.arange(5)

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, squeeze=True)
ax[0][0].plot(x, np.sqrt(x))   # left top
ax[0][1].plot(x, x)            # right top
ax[1][0].plot(x, -x+5)         # left bottom
ax[1][1].plot(x, np.sqrt(-x+5))# right bottom

plt.show()
```
